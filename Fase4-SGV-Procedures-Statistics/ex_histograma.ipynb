```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"ocorrencias_sac_v2.csv")
```

```python
df.columns = df.columns.str.strip()

df["data_ocorrencia"] = pd.to_datetime(df["data_ocorrencia"])
```

```python
frequencias = df["data_ocorrencia"].value_counts().value_counts().sort_index()

distribuicao_prob = (frequencias / frequencias.sum()) * 100

fig, axs = plt.subplots(2, 1, figsize=(12, 8), constrained_layout=True)

# Histograma de frequências absolutas
axs[0].bar(frequencias.index, frequencias.values, color='skyblue')
axs[0].set_title("Frequência de Ocorrências Diárias no SAC")
axs[0].set_xlabel("Número de Ocorrências por Dia")
axs[0].set_ylabel("Número de Dias (Frequência Absoluta)")
axs[0].tick_params(axis='x', rotation=45)

# Histograma de distribuição de probabilidade
axs[1].bar(distribuicao_prob.index, distribuicao_prob.values, color='orange')
axs[1].set_title("Distribuição de Probabilidade de Ocorrências Diárias")
axs[1].set_xlabel("Número de Ocorrências por Dia")
axs[1].set_ylabel("Probabilidade")
axs[1].tick_params(axis='x', rotation=45)

plt.show()
```
